## Problem 2

```python
def compute_sorori_shinzaemon(day=100):
    """
    A function that verifies the anecdote of rice by Sorori Shinzaemon. Calculate the number of rice grains per day.
    Parameteres
    --------------
    day: int
        Number of days to get rice (default : 100)
    Returns
    --------------
    list_n_grains : list
        List of the number of rice you get one day (the length of the list is equal to day)
    list_total_grains : list
        List of the total number of rice you will receive by a certain day (the length of the list will be equal to day)
    """
    # Write the code here. pass is an instruction that means do nothing.
    list_n_grains = [(2**current_day) for current_day in range(day)]
    list_total_grains = [sum(list_n_grains[:current_day+1]) for current_day in range(day)]
    return list_n_grains, list_total_grains
```

```python
day = 10
list_n_grains, list_total_grains = compute_sorori_shinzaemon(day)
print("The number of rice you get {} days is {}".format(day,list_n_grains))
print("The total number of rice you will receive by a certain day of {} days is {}".format(day,list_total_grains))
```

```python
plt.title("The change in the number of rice grains")
plt.xlabel("number of days")
plt.ylabel("number of rice grains")
plt.plot(list_n_grains, 'r--') 
plt.plot(list_total_grains, 'b:') 
plt.show()
```
